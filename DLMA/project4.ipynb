```python
import tensorflow as tf
from tensorflow import keras
import random
import numpy as np
import time
```

```python
tf.config.list_physical_devices('GPU')
```

```python
x1_train=[]
x2_train=[]
y_train=[]
```

```python
m = 1000
n = 100
```

```python
for i in range(m):
    x1_train.append(random.uniform(-2, 2))
    x2_train.append(random.uniform(-2, 2))
    if x1_train[-1]*x1_train[-1] > x2_train[-1]:
        y_train.append(1)
    else:
        y_train.append(0)
```

```python
x_train = np.array([x1_train,x2_train])
y_train = np.array(y_train)
```

```python
x_train = x_train.transpose()
```

```python
x_train
```

```python
y_train = y_train.reshape(1000,1)
```

```python
x1_test=[]
x2_test=[]
y_test=[]
```

```python
for i in range(n):
    x1_test.append(random.uniform(-2, 2))
    x2_test.append(random.uniform(-2, 2))
    if x1_test[-1]*x1_test[-1] > x2_test[-1]:
        y_test.append(1)
    else:
        y_test.append(0)
```

```python
x_test = np.array([x1_test,x2_test])
y_test = np.array(y_test)
```

```python
x_test = x_test.transpose()
y_test = y_test.reshape(100,1)
```

```python
def create_model():
    model = keras.Sequential([
    keras.layers.Dense(3, activation='sigmoid'), #input_shape=(2,)),
    keras.layers.Dense(1, activation='sigmoid')
    ])
    return model
```

## table 1

```python
model = create_model()
```

```python
opt_sgd = keras.optimizers.SGD(learning_rate=0.3)
```

```python
model.compile(optimizer=opt_sgd, loss=tf.keras.losses.BinaryCrossentropy(), metrics=['accuracy'])
```

```python
model.fit(x_train, y_train, batch_size = 1000,epochs=1000)
```

```python
model.evaluate(x_test,y_test)
```

```python
model_2 = create_model()
```

```python
model_2.compile(optimizer=opt_sgd, loss = tf.keras.losses.MeanSquaredError(), metrics=['accuracy'])
```

```python
model_2.fit(x_train, y_train, batch_size = 1000, epochs=1000)
```

```python
model_2.evaluate(x_test,y_test)
```
